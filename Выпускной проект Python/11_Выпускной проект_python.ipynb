# Задача
У интернет-магазина товаров для дома есть проблемы с продажами. Нужно выяснить какие заказчики имееют наибольший потенциал с точки зрения увеличения продаж. Нужно выявить и сгруппировать заказчиков в похожие классы, и проанализировать, какие классы приносят большую прибыль. Далее планируется сделать адресную рассылку с интресными предложениями наиболее перспективным заказчикам. 

# Шаг №1 - загружаем данные.
Набор данных включает следующие поля:

- `date` — дата заказа;
- `customer_id` — идентификатор покупателя;
- `order_id` — идентификатор заказа;
- `product` — наименование товара;
- `quantity` — количество товара в заказе;
- `price` — цена товара.

# Шаг №2 - предобработка данных.
- привести данные к нужным типам;
- проверить на явные и неявные дубликаты;

# Шаг №3 - исследовательский анализ данных.
- изучить средние значения и стндартные отклонения методом describe();
- разбить товары на категории;
- исследовать влияние сезонности на продажи в целом и в разрезе категорий товаров;
- исследовать куммулятивные данные по количеству заказов и выручке за представленный период времени в разрезе групп товаров;
- исследовать распределение выручки по группам товаров.

# Шаг №4 - делаем кластеризацию заказчиков
- Кластеризацию будем проводить по переменным: 'order_id', 'quantity', 'price' и 'revenue' (выручка);
- Построим матрицу расстояний функцией linkage() на стандартизованной матрице признаков и нарисуем дендрограмму. На основании полученного графика предположим, какое количество кластеров можно выделить;
- Обучим модель кластеризации на основании алгоритма K-Means и спрогнозируем кластеры клиентов. Число кластеров определим по итогам исследования дендрограммы;
- Построим распределение количества одноименных товаров в одном заказе, объема выручки и распределение цены товара для кластеров.

# Шаг №5 - проверим гипотезы.
1. Цены товаров в кластерах равны;
- H0 - цены на товары в среднем в кластерах равны.
- H1 - цены на товары в кластерах разные, значит мы правильно разделили заказчиков
2. Выручки пользователей из разных кластеров равны.
- H0 - средняя выручка в кластерах равна.
- H1 - средняя выручка в кластерах разная, значит мы правильно разделили заказчиков

```python
# стандартизируем данные
X = data_km[['revenue_sum', 'quantity_sum', 'price_sum', 'orders_cnt']]
sc = StandardScaler()
X_sc = sc.fit_transform(X)
linked = linkage(X_sc, method = 'ward')
```

```python
# визуализируем таблицу связок linked
plt.figure(figsize=(20,15))
dendrogram(linked, orientation='top')
plt.title('Кластеры')
plt.show()
```

- H0 - статистической разницы между ценами на товары в разных кластерах нет

## Рекомендации
- 0 кластер - это закзачики сделавшие по одному закзазу, их больше всех, но они не возвращаются, можно сделать по ним общую рассылку с рекламой, что бы не забывали.
- 1 кластер - все заказчики этого кластера делают более одного заказа и можно отнести их к лояльным. Они интересуются всеми популярными группами товаров, но в меньшей степени Тележками, Напольными покрытиями и посудой. Им можно предложить рассмотреть товары из этих групп, чтобы расширить номенклатуру их интересов.
- 2 кластер - с одной стороны заказчики этого кластера приносят мало денег. Но есть среди них 21 лояльных покупателей, которые в основном берут Цветы и Рассаду и являются сезонными покупателями. Средний чек у них довольно внушительный, поэтому перед дачным сезоном можно предложить им рассмотреть товары более дорогих сегментов товаров.
- 3 кластер - в этом кластере 67% лояльных клиентов. Вообще все клиенты из этого кластера склонны к покупкам дорогих товаров. Поэтому нужно сделать рассылку по всем клиентам этого сегмента. Это позволит попробовать повторно привлеч тех, кто сделал только один заказ. А остальные тоже может чем то заинтересуются.
